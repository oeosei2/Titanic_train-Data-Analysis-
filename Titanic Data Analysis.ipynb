# DATA VISUALIZATION

```python
data['Survived'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Survived Passengers')
plt.show()
print('0=Death', '1=Survived')
```

```python
data['Survived'].value_counts().plot.bar(color=['r','g'])
plt.title('Distribution of Survived Passengers')
plt.xlabel('Survived')
plt.ylabel('Number of Passengers')
plt.show()
print('0=Death','1=Survived')
```

```python
data['Sex'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Distribution of Sex by Passengers')
plt.show()
```

```python
sns.heatmap(data.corr())
plt.show()
print('The heatmap show the relationship between the features')
```

```python
data['Pclass'].value_counts().plot.bar(color=['violet','blue','brown'])
plt.title('Distribution Of Pclass by Passengers')
plt.xlabel('Pclass')
plt.ylabel('Number of Passengers')
plt.show()
print('1=1st class', '2=2nd class', '3=3rd class')
```

```python
bins=[0,15,30,45,60,75,90]
data['age']=pd.cut(data['Age'],bins)
sns.countplot(data['age'],hue=data['Survived'])
plt.title('Distribution of age by survived passengers')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.countplot(data['age'],hue=data['Sex'])
plt.show()
```

```python
sns.set_style('whitegrid')
sns.violinplot(data['Survived'],data['Age']).set_title('Distribution of Age according to Passengers Survived')
plt.show()
```

```python
plt.figure(figsize=(4,6))
sns.set_style('whitegrid')
sns.pairplot(data,hue='Survived')
plt.show()
```

```python
plt.boxplot(data['Survived'])
plt.show()
```

```python
sns.set_style('whitegrid')
sns.countplot(data['Survived'],hue=data['Sex'])
plt.show()
```
